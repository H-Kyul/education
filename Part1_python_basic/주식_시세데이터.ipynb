```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import time
pd.set_option('display.max_columns',None)
pd.set_option('display.max_row',None)


dfs = []
code = ['185750','005930','035720','051910','063160',
        '006400','068270','068290','005380','006980']

for item in code:
    #print(item)
    for i in range (1,6):
        url = 'https://finance.naver.com/item/sise_day.nhn?code='+str(item)+'&page='+str(i)
        headers = {'User-Agent':'Mozilla/5.0'}
        r = requests.get(url,headers=headers)
        r.text
        time.sleep(0.1)
        if i==1:
            data=pd.read_html(r.text)      
        elif i==2:
            temp = pd.read_html(r.text)
            data = data[0].append(temp[0])
        elif i>2:
            temp=  pd.read_html(r.text)
            data = data.append(temp[0])
    df = pd.DataFrame(data)
    df = df.dropna(axis=0)
    df['종목코드'] = item
    df.reset_index(inplace=True)

    df=df[['종목코드','날짜','종가','전일비','시가','고가','저가','거래량']]
    dfs.append(df)        

    

        
dfs

```
