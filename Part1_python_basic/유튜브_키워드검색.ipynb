# <<실습 1.>> #
유튜브 검색어 입력하고 결과 확인

```python
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
import time 
driver = webdriver.Chrome('C:/Users/gyulll/Downloads/chromedriver.exe')
search = input('검색어:')
youtube = 'https://www.youtube.com'
url = youtube+'/results?search_query='+search

driver.get(url)
time.sleep(10)
html = driver.page_source
soup = BeautifulSoup(html)

video_list = []
data = soup.select('a#video-title')
for i in range(len(data)):
    a = data[i].get('aria-label').split()
    chk1 = a.index('전'); chk2 = a.index('게시자:')
    date = a[chk1-1]+a[chk1]
    channel ="".join(a[chk2+1:chk1-1])
    ViewNum = a[-1]
    video_url = youtube+data[i].get('href')
    title = data[i].text.strip()
    video_list.append([channel,title,video_url,ViewNum,date])

df = pd.DataFrame(video_list,columns = ['Channel','Title','Url','ViewNum','Date'])
df
```

# <<실습 2.>> #
유튜브 채널 재생목록 출력

```python
playlist = []
a = soup.select('ytd-playlist-renderer')
for i in range (len(a)):
    plt_ch = a[i].find('yt-formatted-string',{'class':"style-scope ytd-channel-name complex-string"}).text
    plt_tt = a[i].find('span',{'class':"style-scope ytd-playlist-renderer"}).text.strip()
    plt_url = youtube+a[i].find('a')['href']
    video_num = a[i].find('yt-formatted-string',{'class':'style-scope ytd-thumbnail-overlay-side-panel-renderer'}).contents[0]
    playlist.append([plt_ch,plt_tt,plt_url,video_num])

plt_df = pd.DataFrame(playlist,columns=['Channel','Title','URL','VideoNum'])
plt_df
```
